```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import math
```

```python
# Load the TSV file, using header=None since there is no header row
df_train = pd.read_csv('OliveOil_TRAIN.tsv', sep='\t', header=None)
df_test = pd.read_csv('OliveOil_TEST.tsv', sep='\t', header=None)
```

```python
#  Separate labels and data of train 
labels_train = df_train.iloc[:, 0].values        
data_train = df_train.iloc[:, 1:]     
print("train Labels shape:", labels_train.shape)
print("train Data shape: ", data_train.shape)
```

```python
#  Separate labels and data of test 
labels_test = df_test.iloc[:, 0].values        
data_test = df_test.iloc[:, 1:]     
print("test Labels shape:", labels_test.shape)
print("test Data shape: ", data_test.shape)
```

```python
#  Verify there are no missing values
print("\nAny missing values in the train dataset?", data_train.isnull().any().any())
print("\nAny missing values in the test dataset?", data_test.isnull().any().any())
```

```python
plt.figure(figsize=(25, 4))
count = 0
for i in range(len(labels_train)):
    if labels_train[i] == 1:
        print(f"Plotting sample idx = {i}")
        plt.plot(data_train.iloc[i].values,label=f"idx {i}")
        count += 1
        if count == 5:
            break
plt.title("sample graph of dataset")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend(loc="upper right", fontsize="small")
plt.grid(True)
plt.show()
```

```python
# Extract two time series 
series1 = data_train.iloc[0].values
series2 = data_train.iloc[5].values
series3 = data_train.iloc[14].values
series4 = data_train.iloc[18].values

# Plot the first two time series
first_label = labels_train[0]
second_label = labels_train[5]
third_label = labels_train[14]
fourth_label = labels_train[18]

fig = plt.figure(figsize=(8, 4.5), dpi=600)
plt.plot(series1, label=f"Series 1 (Label: {first_label})", linewidth= 0.5)
plt.plot(series2, label=f"Series 2 (Label: {second_label})", linewidth= 0.5)
plt.plot(series3, label=f"Series 3 (Label: {third_label})", linewidth= 0.5)
plt.plot(series4, label=f"Series 4 (Label: {fourth_label})", linewidth= 0.5)
plt.title("Time Series Samples from Olive Oil Dataset")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig("OliveOil_graph.pdf", bbox_inches='tight')
plt.show()
```
